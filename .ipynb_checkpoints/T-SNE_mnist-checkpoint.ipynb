```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.style as style
style.use('seaborn')
%matplotlib inline
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
x_train = x_train.reshape(x_train.shape[0], -1)
print(x_train.shape)
```

```python
from sklearn.manifold import t_sne
sk_tsne = t_sne.TSNE(n_components=2, perplexity=30, n_iter=5000, init='random',verbose=1)
embedded_tsne = sk_tsne.fit_transform(x_train[:1000])
```

```python
df = pd.DataFrame()
df['x'] = embedded_tsne[:, 0]
df['y'] = embedded_tsne[:, 1]
df['label'] = y_train[:1000]
plt.figure(figsize=(12, 8))
ax = sns.scatterplot(x='x',y='y', hue='label',data=df, palette=sns.color_palette("hls", n_colors=10))
ax.xaxis.label.set_visible(0)
ax.yaxis.label.set_visible(0)
```
